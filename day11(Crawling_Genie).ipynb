```python
# pip install selenium
from selenium import webdriver  #파이썬으로 웹 브라우저 컨트롤

from bs4 import BeautifulSoup as bs # 분석을 용이하게 정제하기 위해 BeautifulSoup

```

```python
driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://www.genie.co.kr/chart/top200')
```

```python
html = driver.page_source
soup = bs(html)
```

```python
len(soup.select('tr.list'))
soup.select('tr.list')
```

```python
songs=soup.select('tr.list')
songs[0]
```

```python
print(songs[0].select('a.artist')[0].text)
print(songs[0].select('a.title')[0].text.strip())
```

```python

rank=1
song_data=[]
songs = soup.select("tr.list")

for song in songs:
    singer = song.select('a.artist')[0].text
    title = song.select('a.title')[0].text.strip()
    print(rank,singer,title)
    song_data.append([rank,singer,title])
    rank=rank+1
print(song_data)
```

```python
import pandas as pd           #DataFrame
df = pd.DataFrame(song_data, columns = ['순위','타이틀','가수'])
df
```

```python
df.to_excel("Genie.xlsx",index=False)
```
