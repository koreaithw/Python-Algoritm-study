```python
---------------------------단순 삽입법---------------------------

삽입하려는 데이터를 변수를 따로 하나 준비하여 대입한다.
변수의 데이터를 이미 정렬된 데이터와 순서대로 비교한다.
변수의 데이터보다 작은 데이터가 발견되면 그 뒤의 요소에 변수의 데이터를 삽입한다.

올바른 순서가 되도록 하나씩 데이터를 삽입한다.
단순 삽입법은 삽입정렬, 기본 삽입법, 삽입법 이라고 한다.

0 1 2 3 4 index
5 3 4 1 2 data

인덱스0을 정렬된 공이라고 가정한다.

5 3 4 1 2
-             정렬된 공이라고 가정

정렬이 안된 공들 중 첫번째 즉 3을 정렬된 공들중 올바른 순서가 되도록 삽입한다

3 5 4 1 2
0번쨰 같의 앞에는 없기 때문에 5를 하나뒤로 이동하고 3인동을 0번째 칸에 넣는다
3과5는 정렬된 상태가 되었다

다시 정렬이 안된 공들 중 첫번째 즉 4를 정렬된 공들중 올바른 순서가 되도록 삽입한다

이 시점에서 '정렬된공'은 0번칸 3과 1번칸 5사이가 되므로 3의뒤 5의 앞이다
4를 5앞에 삽입한다

3 4 5 1 2

다시 정렬이 안된 공들중 첫번째 즉 1을 정렬된 공들중 올바른 순서가 되도록 삽입한다
정렬된 공은 0번칸 3, 1번칸 4 2번칸은 5이다 올바른 위치는 3의 앞이므로 1을 3의 앞에 삽입한다

1 3 4 5 2

마지막으로 2를 정렬된 공들 중 올바른 순서가 되도록 삽입한다.
1 2 3 4 5
```

```python
---------------------------단순 삽입법---------------------------
data=[5,3,4,1,2]
def insertion(data):
    for i in range(1,len(data)):
        while i>0 and data[i]<data[i-1]:
            data[i],data[i-1]=data[i-1],data[i]
            i-=1
    return data
print(insertion(data))
```

```python
---------------------------에라토스 테네스의 체(소수 찾기 알고리즘)---------------------------

소수를 찾아내는 알고리즘(특정 범위의 수 중에서 소수를 찾는 알고리즘)
소수는 2이상의 정수와 1과 그 수 자신으로만 나눌수 있는 수

소수인지 아닌지를 구분하는 것은 의외로 아주 어렵다.
소수는 규칙성이 업기 때문에 간격이 불규칙하고 랜덤하다.즉 소수 

하나하나 그 수보다 작은 숫자로 나누어 보고 나눌수 있으면 소수가 아닌것으로 판단 할 수 있다.
이 방법은 비효율적이고 수의 범위가 커지면 어마어마하게 많은 시간이 소요된다.

고대 그리스의 에라토스테네스는 위와 같은 방법을 개선하여 소수를 효율적으로 발견하는 방법을 고안했다.

에라토스테네스의 체는 '그 수의 제곱근보다 작은 소수의 배수를 없에면 남은수가 소수다.'
라는 생각을 바탕으로 소수를 찾는 방법을 말한다.

예를 들면 100 이하의 소수를 모두 찾고 싶으면 가장먼저 100의 제곱근 이하의 소수를 선택한다,.
100의 제곱근  =10이므로 10이하의 모든 소수 = 2,3,5,7 네개 이다.

먼저 2부터 100까지 중 2이상의 2의 배수를 모두 삭제한다.

나누어 지는 수는 모두 소수가 아니므로 즉, 제곱근 이하의 어느 소수로도 나누어 떨어지지 않고 끝까지 
체에 남아있는 수가 소수이다.
```

```python
---------------------------에라토스 테네스의 체 알고리즘---------------------------
1. 어떤 수 이하의 모든 정수 데이터를 준비
2. 어떤 수의 제곱근 보다 작은 소수의 배수를 제거한다.
3. 마지막까지 남은 수를 출력한다.
```

```python
---------------------------에라토스 테네스의 체 알고리즘---------------------------
p=int(input("정수입력>"))
for i in range(1,p+1):
    count=0
    for k in range(1,i+1):
        if i%k==0:
            count+=1
    if count==2:
        print(i,end=" ")
```

```python
---------------------------유클리드 알고리즘---------------------------
두 수의 나눗셈을 반복하여 최대 공약수를 구한다.

12와 6의 최대 공약수 
12 % 6 = 0 나머지가 없어 딱 나누어 떨어지는데 사용된 수가 최대 공약수.

나머지를 구하는 나눈셈의 산수 연산자 %를 사용한다.
딱 나누어 떨어지지 않을때에 나눗셈을 반복하는 처리를 반복 구조로 표현한다.
반복구조에는 선판정형과 후판정형이 있다.
선 판전형 구조는 반복처리에 들어가기 전에 조건식으로 판정을 실시한다.
후 판정형 구조는 반복 처리 후에 조건식으로 판정을 실시한다.(do while문)자바
```

```python
---------------------------유클리드 알고리즘---------------------------
r1 = int(input("r1="))
r2 = int(input("r2="))
r=0
if r1>r2:
    while r2!=0:
        r=r1%r2
        r1,r2=r2,r
    print("gcd=",r1)
elif r1<r2:
    while r1!=0:
        r=r2%r1
        r2,r1=r1,r
    print("gcd=",r2)
    
```

```python
---------------------------빅 오 표기법---------------------------
알고리즘이 얼마나 빠른지를 표시하는 특별한 방법

                                            탐색
                   단순탐색                                              이진탐색
     100개         100밀리초                                          7밀리
     10000개       10초                                              14밀리
     100000000개   11일                                              32밀리
원소의 갯수가 증가해도 이진 탐색 시간은 얼마 늘어나지 않는다
하지만 단순 탐색의 시간은 어마어마하게 증가 된다.
원소의 갯수가 커질 수록 이진 탐색은 훨씬 빨라지게 된다.

빅오 표기법은 알고리즘이 얼마나 빠른지만 말해준다.

단순 탐색은 원소를 하나하나 확인하지 때문에 n번 연산을 해야한다.
빅오 표기법으로는 O(n)이다. 빅오 표기법은 속도를 시간단위로 세지 않는다.
빅오 표기법은 연산 횟수를 비교하기 위한 개념이다.
단순 탐색의 실행 시간은 O(n) 로 나타낸다.

O(log n) 이진 탐색  ---------아래로 갈 수록 느려짐
O(n) 선형 시간 , 단순 탐색  
O(n * log n) 
O(n^2) 선택 정렬, 느린 정렬 알고리즘
O(n!) 외판원 문제, 정말 느린 알고리즘

알고리즘의 속도는 시간으로 측정하지 않는다.
알고리즘의 시간은 어ㄸ허게 증가하는 가로 측정한다.
알고리즘의 시간은 빅오표기법으로 나타낸다.
```
