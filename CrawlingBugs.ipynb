```python
#pip install requests  #에러나면 install
import requests  #웹 페이지 연결 모듈
```

```python
from bs4 import BeautifulSoup as bs # 읽어온 웹 페이지를 분석하기 쉽게 정제 (BeautifulSoup로 정제해야 분석이 쉽다)
```

```python
import pandas  #데이터 분석 모듈
```

```python
html = requests.get('https://music.bugs.co.kr/chart')  #웹 페이지 코드를 그냥 글자로 긁어옴,코드의 의미가 사라짐
# print(html.text)
```

```python
soup = bs(html.text) #BeautifulSoup 을 통해 데이터를 분석하기 용이하게 정제한다. 진짜 HTML
# print(soup)
```

```python
len(soup.select('tr')) #tr태그가 붙은 부분들을 모두 추출 할때 사용하는 함수 SELECT
```

```python
len(soup.select('tbody > tr')) #tbody밑에 있는 tr태그가 붙은 부분들만 추출(범위 축소)
```

```python
len(soup.select('table > tbody > tr')) # table밑의 tbody밑의 tr태그 추출
```

```python
len(soup.select('table.byChart > tbody > tr'))    #.은 클래스를 의미 list trackList byChart에서 빈칸이있을경우 맨 마지막만 사용
```

```python
song_data=[]
rank =1

songs =soup.select('table.byChart > tbody > tr')  #100개 추출

for song in songs:
  title = song.select('p.title > a')[0].text
  singer = song.select('p.artist > a')[0].text
  song_data.append([rank,title,singer])
  rank=rank+1
song_data
```

```python
import pandas as pd           #DataFrame
df = pd.DataFrame(song_data, columns = ['순위','타이틀','가수'])
df
```

```python
df.to_excel('bugs.xlsx',index=False)
```
