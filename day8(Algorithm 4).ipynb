```python
#################
단순 교환법 Bubble sort
#################
-인접한 데이터를 교환하는 처리를 반복한다,

0    1    2    3    4 index
5    3    4    1    2 data

5    3    4    1    2  이미 정렬이 된 상태임에 따라 정렬할 필요가 없다.
               -    - 
    
5    3    4    1    2  정렬이 되지 않은 상태임에 따라 교환한다.
          -    - 
    
5    3    1    4    2  정렬이 되지 않은 상태임에 따라 교환한다.
     -    -
    
5    1    3    4    2  정렬이 되지 않은 상태임에 따라 교환한다.
-    -

1 |  5    3    4    2  "1"은 정렬이 완전히 끝난 상태로 확정되었다.

1 |  5    3    4    2
               -    -
    
1 |  5    3    2    4
          -    -
    
1 |  2    5    3    4
     -    -
    
1    2 |  5    3    4  "2"는  정렬이 완전히 끝난 상태로 확정되었다.


1    2 |  5    3    4
               -    -
    
1    2 |  5    3    4
          -    - 
    
1    2    3 |  5    4  "3"는  정렬이 완전히 끝난 상태로 확정되었다.

1    2    3 |  5    4
               -    -
    
1    2    3    4    5   정렬이 완전히 끝난 상태로 확정되었다.
```

```python
단순 교환법 bubble sort 알고리즘
오른쪽 끝부터 순서대로 인접한 공 2개를 오름차순으로 정렬한다.
왼쪽 끝 칸부터 순서대로 들어갈 공을 확정시켜 나간다.

0    1    2    3    4 index
5    3    4    1    2 data


오른쪽 끝 인접한 공 2개
오른쪽 요소 i    arr[i]
왼쪽 요소 i-1    arr[i-1]

            정렬이 이미 되어있는 상태인지 여부 확인 >> 

                "왼쪽요소가 오른쪽요소보다 작은가?"
                        arr[i-1]<arr[i]
```

```python
여기까지는 순조롭게 비교와 교환 처리가 진행된다,.
마지막에 i-1이 0 이 되고 i 가 1 이 되면
이미 확정된 arr[0]과 arr[1]을 비교하게 된다. 이것은 쓸데없는 낭비가 된다.
그러면 어디를 수정해야 할까..  i>0 이 조건식이 문제이다.
따라서 
arr[1]을 확정하는 반복 처리부터는 i>0를 i>1로 바꾸어야 한다.
변수 k로 표현하자 k의 초기값을 0 부터 시작 i>k
```
