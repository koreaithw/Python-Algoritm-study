```python
1 알고리즘이란...
-문제나 과제를 해결하기 위한 처리 절차를 
구체적인 순서로 표현한 아이디어 또는 생각
-알고리즘 자체는 일상생활에서도 많이 사용하고 있다.
-요리의 레시피, 음악의 악보, 가전제품 사용설명서 등..
-문자, 기호, 일러스트 등으로 나타낸다.

2 알고리즘과 프로그램
-알고리즘을 프로그래밍 언어로 기술하면 프로그램이 된다.
-프로그래밍 언어(파이썬, 자바, 씨..)는 컴푸터에 지시하기 위한 인공 언어이다.

3 프로그램 작성에서의 알고리즘
-프로그램 작성은 기획, 설계(*), 프로그래밍, 디버그로 진행한다.
-알고리즘으느 설계 단계에서 필요하다.

4 좋은 알고리즘이란??
-알기쉽다 (쉬운이해)
-속도가 빠르다.
-효율적이다 (메모리 사용이 적다)
-재이용(범용성)이 쉽다.(객체지향프로그래밍)

5 왜 알고리즘 공부를 하나
-좋은 프로그램을 만들기 위해(빠르고, 효율적, 범용성)
-프로그램의 좋고 나쁨을 판단하기 위해
-프로그램 작성 과정 전체를 효율화 하기위해
-프로그램  기술을 향상시키기 위해

6 알고리즘의 조건
-정확한 결과를 얻을 수 있어야 한다.
    (알고리즘은 문제나 과제를 해결하기 위한 절차)
-반드시 종료 되어야 한다.
    (중간처리나 절차가 잘못 만들어지면 영원히 끝나지 않는다.(무한루프))
    
7 알고리즘의 기본 3가지
-순차구조 : 처음부터 순서대로 처리하는 절차
    1. 수퍼마켓에 간다.
    2. 라면을 산다.
    3. 집으로 돌아온다.
    실행을 원하는 처리를 위로부터 순서대로 작성한다.
    알고지름 중에서도 가장 많이 사용된다.
    
-선택구조 : 조건식으로 판단해 실행할 절차를 전환하는 구조
    
-반복구조 : 조건을 만족하는 동안 같은 처리를 반복하는 절차
```

```python
def sum_all(start,end):
    output =0
    for i in range(start,end+1):
        output=output+i
    return output    
print("0에서 100까지의 합 :",sum_all(0,100))
print("0에서 100까지의 합 :",sum_all(0,1000))
print("0에서 100까지의 합 :",sum_all(50,100))
print("0에서 100까지의 합 :",sum_all(500,1000))
```

```python
#반복문으로 팩토리얼 구하기
def facto(n):
    output=1
    for i in range(1,n+1):
        output = output*i
    return output

print("3! :",facto(3))
print("4! :",facto(4))
print("5! :",facto(5))
```

```python
#재귀함수로 팩토리얼 구하기
def facto(n):
    if n==0:
        return 1
    else:
        return n*facto(n-1)
    
print("3! :",facto(3))
print("4! :",facto(4))
print("5! :",facto(5))

# 개발자들 사이에서는 재귀함수를 절대 사용하지 말자 라는 의견도 있다.
# 하지만 가독성이 좋고, 
```

```python
#피보나치 수열
def fibo(n):
    if n==1 or n==2:
        return 1
    else:
        return fibo(n-2)+fibo(n-1)
    
print("fibonacci(1)",fibo(1))
print("fibonacci(2)",fibo(2))
print("fibonacci(3)",fibo(3))
print("fibonacci(4)",fibo(4))
print("fibonacci(5)",fibo(5))
```

```python
def add_many(*args):
    result =0
    for arg in args:
        result += arg
    return result
print(add_many(1,2,3,4,5))
print(add_many(1,2,3))
```

```python
def add_mul(choice, *args):
    if choice == "add":
        result=0
        for i in args:
            result += i
  
    elif choice == "mul":
        result=1
        for i in args:
            result = result * i
    return result
print(add_mul("add",1,2,3,4,5))
print(add_mul("mul",1,2,3))
```

```python
# 함수의 결과는 언제나 하나이다  #여러개 일경우 튜플이 한개

def add_and_mul(a,b):
    return a+b,a*b
result1,result2 = add_and_mul(3,4)
print(result1)
print(result2)
```
