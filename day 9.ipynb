```python
n=input("입력값>>")
a=""
b=""
for i in n:
    if i.isdigit():
        b=b+i    
    else:
        a=a+i
print("문자:",a)     
print("숫자:",b)
```

```python
문자와 숫자의 개수를 포함하여 출력
```

```python
n=input("입력값>>")
a=""
b=""
c=0
d=0
for i in n:
    if i.isdigit():
        c=c+1
        b=b+i    
    else:
        d=d+1
        a=a+i
print(f"문자{c}: {a}")     
print(f"숫자{d}: {b}")
```

```python
ip=input("입력값>>")
cs=0
cn=0
s=""
n=""
for i in ip:
    if "A"<=i<="Z":
        cs=cs+1
        s=s+i
    elif 'a'<=i<='z':
        cs=cs+1
        s=s+i
    elif '0'<=i<='9':
        cn=cn+1
        n=n+i
        
print(f"문자{cs}: {s}")     
print(f"숫자{cn}: {n}")
```

```python
################################################################################################################
########################  알고리즘 복습  ########################################################################
################################################################################################################

- 알고리즘은 문제나 과제를 해결 하기 위한 처리 절차를 구체적인 순서로 표현한 아이디어나 생각.
- 알고리즘 표현 방법은 수도코드(pseudo code), 순서도(flow chart)

-프로그램 작성순서
-> 기획 - 설계(알고리즘) -프로그래밍 - 디버그 - 문서작성

*좋은 알고리즘
*알기쉽고, 속도가 빠르고, 효율적이고(메모리를 적게사용), 재이용이 쉽다.

왜 알고리즘인가?
-좋은 프로그램을 만들 수 있고, 좋은 프로그램인지 판단할 수 있고,
-프로그램 작성을 효율적으로 할 수 있고 프로그래밍 기술을 향상시키는데 큰 도움이 된다.

---알고리즘의 조건
-결과를 얻을 수 있어야 하고 반드시 종료 되어야 한다.
1. 순차구조
2. 선택구조
3. 반복구조

-순서도 flow차트, 흐름도
-프로그래밍 언어를 사용하지 않고 알고리즘을 기술하는 방법
처리와 절차를 시각적으로 표현한다.
자신이 생각하고 있는 알고리즘이 올바른지, 처리와 절차에 문제가 없는지를
눈으로 확일 할 수 있고 다른 사람에게도 쉽게 전달 할 수 있는 특징이 있다.

~!!!!~~ 주의사항
처리기호나 판단기호가 많아지면 흐름이 복잡하게 꼬인다.
1. 입력은 반드시 위에서 들어가도록 만든다.
2. 판단기호로 부터 나오는 흐름선에는 yes/no를 확실하게 기입한다.
3. 입력선이나 출력선이 없는 처리 기호는 만들지 않는다.
```

```python
유명한 알고리즘?
유명한 알고리즘을 배우는 것은 프로그래밍 기술 향상에 커다란 도움이 된다.

-탐색 ; Linear Esarch, Binary Search, Hash Search
-정렬 : Selection Sort, Bubble Sort, Insertion Sort, Quick Sort
-Eieve of Eratosthenes, Eucild Algorithm

가우스 소거법, 사다리꼴 법칙, 디익스트라 알고리즘, 이분법, 뉴턴법
KMP알고리즘, BM알고리즘
```

```python
이진탐색법 Binary Search

-탐색 대상데이터들이 미리 오름차순 또는 내림차순으로 정렬이 되어있어야만 한다.
-탐색하는 범위를 절반씩 추려나가면서 탐색을한다.

0   1   2   3   4   5   6  index
11  13  17  19  23  29  31 data
                                가운데 인덱스 3을 열어 확인 
                                찾고있는 수가 아니지만 정렬되어 있으므로 
                                인덱스 3의 값이 찾는 수보다 작으면 중안인덱스+1을 head의 인덱스로 지정
                                크면 중앙인덱스 -1을 tail의 인덱스 값으로 지정
                                다시 중앙인덱스 값계산 검색

#############################  이진탐색법 알고리즘    ####################################################
1. 가운데 요소를 선택하는 처리(인덱스의 평균을 계산하여 구한다.)
2.가운데 데이터와 찾는 데이터 비교
3. 탐색범위를 절반으로 줄이는 처리

if 요소의 갯수가 짝수일때는 어떻게 해야할까?
0   1   2   3   4   5   index
11  13  17  19  23  29  data

가운데 요소가 짝수일 때는 가운데 요소가 2개이다
이 경우 아무쪽이나 기준을 잡으면 된다.
소수를 정수화 하는 방법  1.반올림 2.올림 3.버림
# ceil()  올림
# floor() 내림
# round() 반올림
```
