```python
## Binary Search
def binary_search(n,lists):       #binary_search 
    head=0                        # head
    tail=len(lists)-1             # tail
    for i in range(len(lists)):  
        if head<=tail:            # head  tail   
            c=(head+tail)//2      # 
            if lists[c]==n:       #   
                print(f"{c}  ")
                break
            elif lists[c]<n:      #     head     
                head=c+1
            else:                 #     tail     
                tail=c-1
        else:                      #    
            print(".")
            break
    
binary_search(19,[11,13,17,19,23,29,31])
```
