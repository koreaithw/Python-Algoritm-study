```python
유명한 알고리즘....
기본적인 처리 절차를 가지고 있는 좋은 알고리즘이다.
좋은 프로그램을 만들기 위한 사고방식이나 힌트가 많이 포함되어 있다.
유명한 알고리즘을 학습하는 것은 프로그래밍 기술 향상에 커다란 도움이 된다.

탐색 Search
Linear Search (선형)-앞에서부터 순서대로 찾는다(속도가 느리다는 단점 비효율적)
Binary Search (이진)-범위를 절반식 줄여서 찾는다(속도가 빠르다는 장점,단 데이터가정렬이 되어 있어야한다.)
Hash Search   (해쉬)-검색 즉시 찾는다.(필요한것만 찾는다는 장점,단 저장할때 가공해서 저장한다.) 

정렬 Sort
Selection Sort 단순 선택법 - 최소/최대값을 선택하여 앞부터 순서대로 나열한다.
Bubble Sort 단순 교환법 - 옆에 있는 데이터를 교환하면서 자리를 바꾼다.
Insertion Sort 단순 삽입법 -데이터를 올바른 위치에 삽입하면서 자리를 바꾼다.
Quick Sort 퀵 정리 - 기준 데이터를 기반으로 대소 분할을 반복하여 자리를 바꾼다.

에라토스테네스의 체 Sieve of Eratosthenes = 소수 구하는 알고리즘
유클리드 알고리즘 - 최대 공약수를 구하는 알고리즘
```

```python
#Binary Search

#이진 탐색법은 탐색하는 범위를 절반씩 줄여서 탐색하는 방법이다
#데이터들이 먼저 오름차순이나 내림차순으로 정렬되어 있어야만 사용가능.
 
11, 13, 17, 19, 23, 29, 31 >>>> 찾는값은 17

1st step
0   1   2   3   4   5   6    제일 가운데 있는 3번을 열어본다.

#이진 탐색법의 알고리즘
1. 가운데 요소를 선택
2. 가운데 요소의 데이터와 찾는 데이터를 비교
3. 탐색 범위를 절반으로 줄임
```

```python
## Binary Search
def binary_search(n,lists):       #binary_search 함수정의
    head=0                        #리스트의 head
    tail=len(lists)-1             #리스트의 tail
    for i in range(len(lists)):  
        if head<=tail:            #리스트의 head 가 tail보다 클때 까지 계속찾음
            c=(head+tail)//2      #중앙값 설정
            if lists[c]==n:       #중앙값이 원하는 값일 경우
                print(f"{c}번째 요소 일치")
                break
            elif lists[c]<n:      #중앙값이 원하는 값보다 작을 경우 head값을 옮겨 중앙값보다 작은값 모두 없엠
                head=c+1
            else:                 #중앙값이 원하는 값보다 클 경우 tail값을 옮겨 중앙값보다 큰값 모두 없엠
                tail=c-1
        else:                      #원하는 값을 못 찾았을 경우
            print("찾지못했습니다.")
            break
    
binary_search(19,[11,13,17,19,23,29,31])
```

```python
def test():
    print("함수의 첫 줄입니다.")
    try:
        print("try구문 실행되었습니다")
        return  ## 함수에서 빠져나감
    except:
        print("except 구문이 실행되었습니다")
    else:
        print("else 구문이 실행되었습니다")
    finally:
        print("finally 구문이 실행되었습니다")
        
    print("함수의 마지막 줄입니다")   
test()
```

```python
print("프로그램 실행")
while True:
    try:
        print("try 구문실행")
        break  ##while 문에서 벗어남
        print("try구문의 break아래")
    except:
        print("except 구문 실행")
    finally:
        print("finally 구문실행")
    print("while의 마지막줄")
print("프로그램종료")
```
